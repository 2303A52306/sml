```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

df = pd.read_csv('/content/Salary_Data (5).csv')

feature_column = df.columns[0]
target_column = df.columns[-1]

X = df[[feature_column]]
y = df[target_column]

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', label='Best Fit Line')
plt.title(f'Linear Regression: {feature_column} vs {target_column}')
plt.xlabel(feature_column)
plt.ylabel(target_column)
plt.legend()
plt.show()

print(f"Slope (Coefficient): {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
df = pd.read_csv('/content/Salary_Data (6).csv')
feature_column = df.columns[0]
target_column = df.columns[-1]
X = df[[feature_column]]
y = df[target_column]
X_train_20, X_test_20, y_train_20, y_test_20 = train_test_split(X, y,test_size=0.20,random_state=42)
X_train_25, X_test_25, y_train_25, y_test_25 = train_test_split(X, y,test_size=0.25,random_state=42)
model = LinearRegression()
model.fit(X_train_20, y_train_20)
y_pred_20 = model.predict(X_test_20)
model.fit(X_train_25, y_train_25)
y_pred_25 = model.predict(X_test_25)
mse_20 = mean_squared_error(y_test_20, y_pred_20)
mae_20 = mean_absolute_error(y_test_20, y_pred_20)
mse_25 = mean_squared_error(y_test_25, y_pred_25)
mae_25 = mean_absolute_error(y_test_25, y_pred_25)
print(f"Test size 20% - MSE: {mse_20:.4f}, MAE: {mae_20:.4f}")
print(f"Test size 25% - MSE: {mse_25:.4f}, MAE: {mae_25:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
data = {
'Feature1': [1.2, 2.4, 3.5, 4.1, 1.9, 3.7, 2.1, 3.3, 2.8, 4.0],
'Feature2': [2.3, 1.9, 2.7, 3.0, 2.8, 2.5, 3.2, 2.4, 3.1, 3.3],
'Feature3': [3.1, 2.8, 1.5, 3.6, 2.5, 1.9, 2.2, 3.0, 1.8, 2.7],
'Feature4': [4.2, 3.5, 2.9, 4.8, 3.2, 4.0, 4.1, 4.5, 3.6, 4.9],
'Target': [15.6, 13.1, 12.4, 18.2, 14.1, 16.0, 14.5, 17.3, 14.9, 19.2]
}
df = pd.DataFrame(data)
X = df[['Feature1', 'Feature2', 'Feature3', 'Feature4']]
y = df['Target']
plt.figure(figsize=(10, 6))
for column in X.columns:
    plt.scatter(X[column], y, label=column)
plt.xlabel('Features')
plt.ylabel('Target')
plt.legend()
plt.show()
def linear_regression_analysis(test_size):
   X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=test_size,random_state=42)
   model = LinearRegression()
   model.fit(X_train, y_train)
   y_pred = model.predict(X_test)

   mse = mean_squared_error(y_test, y_pred)
   mae = mean_absolute_error(y_test, y_pred)

   return mse,mae

   mse_20, mae_20 = linear_regression_analysis(0.2)
mse_30, mae_30 = linear_regression_analysis(0.3)
print(f'MSE for test size 20%: {mse_20}')
print(f'MAE for test size 20%: {mae_20}')
print(f'MSE for test size 30%: {mse_30}')
print(f'MAE for test size 30%: {mae_30}')
```
